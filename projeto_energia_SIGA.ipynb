Para visualizar as informações das outras fontes, a solução encontrada foi dividir o dataframe por tipo de fonte.

Iniciando pelas usinas hídricas:

```python
hidricas_operando = renovaveis_operando[(renovaveis_operando['DscOrigemCombustivel'] == 'Hídrica')]
hidricas_operando
```

```python
ax = sns.lmplot(data = hidricas_operando, x="MdaGarantiaFisicaKw", y="MdaPotenciaFinalizadaKw")
ax.figure.set_size_inches(12, 6)
ax.fig.suptitle('Potência homologada X Potência registrada - Hídrica', fontsize = 18)
ax.set_axis_labels("Potência registrada", "Potência gerada")
#plt.ylim((-50, 10000000))
plt.ticklabel_format(style='plain')
ax = ax
```

Considerando que o processo de divisão por fonte seria repetido, foi criado uma função para plotar o gráfico das demais fontes:

```python
def dividindo_fontes(fonte, titulo):
  fonte_operando = renovaveis_operando[(renovaveis_operando['DscOrigemCombustivel'] == fonte)]
  fonte_operando
  ax = sns.lmplot(data = fonte_operando, x="MdaGarantiaFisicaKw", y="MdaPotenciaFinalizadaKw") 
  ax.figure.set_size_inches(12, 6)
  ax.fig.suptitle(titulo, fontsize = 18)
  ax.set_axis_labels("Potência registrada", "Potência homologada pelo Ministério das Minas e Energia")
  ax = ax
```

```python
dividindo_fontes('Biomassa', 'Potência homologada X Potência registrada - Biomassa')
```

```python
dividindo_fontes('Eólica', 'Potência homologada X Potência registrada - Eólica')
```

```python
dividindo_fontes('Solar', 'Potência homologada X Potência registrada - Solar')
```

```python
dividindo_fontes('Undi-Elétrica', 'Potência homologada X Potência registrada - Undi-Elétrica')
```

Plot de gráficos da distribuição das fontes de combustível por estado:

```python
def nome_combustivel(combustivel):
  por_combustivel = renovaveis_operando[(renovaveis_operando['NomFonteCombustivel'] == combustivel)]
  ax = sns.displot(data=por_combustivel, x="SigUFPrincipal")
  ax.figure.set_size_inches(12, 6)
  ax.set(title = 'Distribuição da fonte de combustível por estado - {}'.format(combustivel))
  ax.set_axis_labels("Unidade Federativa", "Quantidade de usina")
  ax = ax
```
