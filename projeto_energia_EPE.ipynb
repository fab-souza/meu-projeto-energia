E para finalizar, também havia registros de ‘LGN’, ‘GÁS DE ALTO FORNO COQUE’, ‘GÁS DE ACIARIA’, ‘GÁS DE ENXOFRE E OUTROS’ e ‘OUTRAS Ñ RENOV.’ que foram excluídos para formar o dataset ‘epe2021_renovaveis’, composto por 118 linhas e 19 colunas.

```python
epe2021.iloc[:,59:]
```

```python
epe2021_menos_44_a_54.iloc[:,18:]
```

```python
epe2021_renovaveis = epe2021_menos_44_a_54.drop(columns = epe2021_menos_44_a_54.iloc[:,18:])
epe2021_renovaveis
```

A coluna ‘FLUXO’ substituiu o index numérico, para facilitar a mudança dos valores para o tipo numérico.

```python
epe2021_renovaveis.set_index('FLUXO', inplace=True)
epe2021_renovaveis
```

Com a alteração do dataset finalizado, foi feita a mudança dos valores para o tipo numérico.

```python
epe2021_renov_float = epe2021_renovaveis.iloc[:,:].astype(float)
epe2021_renov_float.head()
```

Primeiramente, o foco consiste no quanto cada fonte produziu durante o período, por isso selecionou-se apenas a primeira linha, ‘PRODUÇÃO’.

```python
producao2021 = epe2021_renov_float.iloc[[0]]
producao2021
```

Ao tentar plotar um gráfico desta última seleção, o resultado foi um pouco diferente de um gráfico comum.

```python
producao2021.plot.bar(title='Produção em 2021');
```

A forma encontrada para plotar um gráfico dessas informações, foi fazer uma transposição do dataset. Assim, cada linha tornou-se referência de uma fonte de energia.

```python
producao2021_T = producao2021.transpose()
producao2021_T
```

Ao tentar plotar um gráfico novamente, este foi o resultado.

Apenas uma observação: a informação referente a ‘Total’ é a soma das fontes Gás industrial Carvão Vegetal, biogás, biomassa e óleos vegetais.


```python
producao2021_T.plot.bar(title='Produção em 2021');
```

Plotando o mesmo gráfico com o Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 7))
plt.title('Produção de energias renováveis em 2021', loc='center')
plt.bar(producao2021_T.index, producao2021_T['PRODUÇÃO'])
# Roacionando o tick (que são os 'índices' na label) em 75°
plt.xticks(rotation=75);
plt.show();
```

Para fins de comparação, também plotei um gráfico com todas as fontes.

```python
producao_geral = epe2021.iloc[[0]]
producao_geral.set_index('FLUXO', inplace=True)
producao_geral.iloc[:,:].astype(float)
producao_geral_T = producao_geral.transpose()
```

Diante o gráfico, em comparação com o petróleo, fontes renováveis como a solar ou biogás, que entre as renováveis já possuíam índices de baixa geração, apresentam-se minúsculas.

Para que não haja confusão, existem dois termos usados para definir a geração de energia:

*   Matriz energética
*   Matriz elétrica

Segundo Grecco, a matriz energética:

"representa um conjunto de fontes de energia disponíveis no país para suprir as demandas energéticas. É por meio dela que será possível captar e distribuir energia para os setores comerciais, industriais e residenciais. Assim, a matriz energética representa a quantidade de energia disponível, bem como a origem dessa energia, se ela é de uma fonte renovável ou não."

Fonte: https://www.gnpw.com.br/matriz-energetica/quais-as-principais-matrizes-energeticas-do-brasil/

A EPE também disponibiliza estas informações no site [BEN Interativo](http://shinyepe.brazilsouth.cloudapp.azure.com:3838/ben/), em que é possível ter acesso a diversos dados, por exemplo verificar a oferta e demanda de energia por fonte, o consumo de energia por setor e análise energética, como o fluxo energético

Já a matriz elétrica:

"[...] é formada pelo conjunto de fontes disponíveis apenas para a geração de energia elétrica. Dessa forma, podemos concluir que a matriz elétrica é parte da matriz energética.

A matriz elétrica brasileira é ainda mais renovável do que a energética, isso porque grande parte da energia elétrica gerada no Brasil vem de usinas hidrelétricas. A energia eólica também vem crescendo bastante, contribuindo para que a nossa matriz elétrica continue sendo, em sua maior parte, renovável."
